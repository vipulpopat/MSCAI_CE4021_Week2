```python
import time

# work out derivative value using following formular
#df(x0) = (f(x0+delta)-f(x0))/delta
def derivative_f1(inde_var):
    delta = 0.001
    derivation_value = (f1(inde_var+delta)-f1(inde_var))/delta
    return derivation_value

def derivative_g(inde_var):
    delta = 0.001
    derivation_value = (g(inde_var+delta)-g(inde_var))/delta
    return derivation_value

def derivative_h(inde_var1, inde_val2):
    delta = 0.001
    derivation_value = (h(inde_var1+delta, inde_val2)-h(inde_var1, inde_val2))/delta
    return derivation_value

def derivative_k(inde_var):
    delta = 0.001
    derivation_value = (k(inde_var+delta)-k(inde_var))/delta
    return derivation_value

#input functions
def f1(x):
    return 3*pow((x), 2)
def g(x):
    return pow((x+8), 2)

def h(a, x):
    return a*pow(x, 3)+ pow(x, 8)/2

def k(x):
    return pow(x, 501)+3*pow(x, 7)+pow(x, 6)/2+pow(x,5)+2*pow(x, 3)+3*pow(x, 2)-1

#Class that save input functions
class CallbackList(list):
    def fire(self, *args, **kwargs):
        for listener in self:
            listener(*args, **kwargs)

def main():
    value = 0
    #append functions to CallbackList
    cb = CallbackList()
    cb.append(derivative_f1)
    cb.append(derivative_g)
    cb.append(derivative_h)
    cb.append(derivative_k)

    #make function map to get input function from string input
    function_map = {'d_f':0, 'd_g':1, 'd_h':2, 'd_k':3}
    print('There are functions "d_f, d_g, d_h, d_k"')
    function_name = input('Select function: ')
    function = cb[function_map[function_name]]

    if function_name == 'd_h':
        a = input('Input a:')
        x = input('Input x:')
        value = function(int(a), int(x))
    else:
        x = input('Input x: ')
        value = function(int(x))
    print('Value is:%.3f'%value)

if __name__ == '__main__':
    main()
```
